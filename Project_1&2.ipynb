```python
%%bash
#--------------------
# Biocomputing Task
# By: Bassa Joshua Samuel
# Date: 02/09/2025
# BASh Basic and Installing Bioinformatics Software on the Terminal
#--------------------

# 1. Print your name
echo "Bassa Joshua Samuel"

#--------------------
# 2. Create a folder titled your name
mkdir Bassa_Joshua_Samuel

#--------------------
# 3. Create another new directory titled 'biocomputing' and change to that directory
mkdir biocomputing && cd biocomputing

#--------------------
# 4. Downloading files
wget https://raw.githubusercontent.com/josoga2/dataset-repos/main/wildtype.fna
wget https://raw.githubusercontent.com/josoga2/dataset-repos/main/wildtype.gbk
wget https://raw.githubusercontent.com/josoga2/dataset-repos/main/wildtype.gbk.1

#--------------------
# 5. Move the '.fna' file to the folder titled my name
mv wildtype.fna ../Bassa_Joshua_Samuel/

#--------------------
# 6. Delete the duplicate '.gbk' file
rm wildtype.gbk.1

#--------------------
# 7. Confirm if the '.fna' file is mutant or wild type ('tatatata' vs 'tata')
cd ../Bassa_Joshua_Samuel

check_mutation () {
    local file=$1
    if grep -q "tatatata" "$file"; then
        echo "Mutant sequence detected in $file"
        grep "tatatata" "$file" > mutant_sequence.txt
        echo "Mutant lines saved to mutant_sequence.txt"
    else
        echo "Wildtype sequence in $file"
    fi
}

check_mutation wildtype.fna

#--------------------
# 8. If mutant, print all matching lines into a new file
grep "tatatata" wildtype.fna > mutant_sequence.txt

#--------------------
# 9. Count number of lines (excluding header) in the '.gbk' file
cd ../biocomputing
grep -v "^LOCUS" wildtype.gbk | wc -l

#--------------------
# 10. Print the sequence length of the '.gbk' file (from LOCUS line)
grep "^LOCUS" wildtype.gbk | awk '{print $3, $4}'

#--------------------
# 11. Print the source organism of the '.gbk' file
grep "SOURCE" wildtype.gbk

#--------------------
# 12. List all the gene names of the '.gbk' file
grep "/gene=" wildtype.gbk

#--------------------
# 13. Clear terminal and print all commands used today
clear
history

#--------------------
# 14. List files in the two folders
echo "Files in Bassa_Joshua_Samuel:"
ls ../Bassa_Joshua_Samuel
echo "Files in biocomputing:"
ls ../biocomputing
```
